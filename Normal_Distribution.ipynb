```python
import pandas as pd
import numpy as np
```

```python
beml_df = pd.read_csv('/content/BEML.csv')
glaxo_df = pd.read_csv('/content/GLAXO.csv')
```

```python
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
# Using date as index using DatetimeIndex
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
# Calculating total gain of stock on perticular day using pct_change() method
beml_df['gain'] = beml_df.Close.pct_change(periods=1)
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods=1)
```

```python
beml_df = beml_df.dropna()
glaxo_df = glaxo_df.dropna()
```

```python
# Ploting gains of both beml and glaxo
sn.distplot(glaxo_df.gain, label = 'Glaxo')
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
```

```python
sn.distplot(beml_df.gain, label = 'BEML')
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
```

```python
# calculating mean and std deviation for gain column
print('Mean : ' ,round(beml_df.gain.mean(),4))
print('Std Deviation : ',round(beml_df.gain.std(),4))
```

```python
print('Mean : ', round(glaxo_df.gain.mean(),4))
print('Std Deviation : ', round(beml_df.gain.std(),4))
```

```python
# Probability of making 2% loss or higherin Beml
stats.norm.cdf(-0.02,loc=beml_df.gain.mean(),scale=beml_df.gain.std())
```

```python
# Probabilty of making 2% or higher gain in Beml
1 - stats.norm.cdf(0.02, loc = beml_df.gain.mean(),scale = beml_df.gain.std())
```

```python
# Probability of making 2% or higher loss in Glaxo
stats.norm.cdf(-0.02, loc = glaxo_df.gain.mean(), scale = glaxo_df.gain.std())
```

```python
# Probability of making 2% or higher gain in Glaxo
1 - stats.norm.cdf(0.02, loc = glaxo_df.gain.mean(), scale = glaxo_df.gain.std())
```
