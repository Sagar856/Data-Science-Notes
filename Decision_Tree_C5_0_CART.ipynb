```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split
# 'train_test_split' splits dataset into train and test set randomly
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.metrics import classification_report
```

```python
iris = pd.read_csv('/content/iris.csv',index_col = 0)
iris.head()
```

```python
iris.rename({'Species':'Iris'}, axis = 1)
```

```python
# Converting Species into numeric from
label_encoder = preprocessing.LabelEncoder()
iris['Species'] = label_encoder.fit_transform(iris['iris'])
```

```python
iris
```

```python
x = iris.iloc[:,0:4]
y = iris['Species']
```

```python
x
```

```python
y
```

```python
iris['Species'].unique()
```

```python
iris['Species'].value_counts()
```

```python
# Spliting dataset into Training and Testing sets

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 40)
```

## Building Decision Tress Classifier using Entropy criteria

```python
model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3)
model.fit(x_train,y_train)
```

```python
# Plot the decision tree
tree.plot_tree(model)
```

```python
# Plotting tree using seaborn
fn = ['sepel lenght (cm)', 'sepel width (cm)', 'petel length (cm)', 'petel width (cm)']
cn = ['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 200)
tree.plot_tree(model, feature_names = fn, class_names = cn, filled = True)
```

```python
# Prediction on test deta
preds = model.predict(x_test)
pd.Series(preds).value_counts() # Getting the count of each category
```

```python
preds
```

```python
# Getting 2 way table to understand the correct and wrong predictions
pd.crosstab(y_test, preds)
```

```python
# Accuracy of model
np.mean(preds==y_test)
```

# CART using Gini criteria

```python
model_gini = DecisionTreeClassifier(criterion = 'gini', max_depth = 3)
```

```python
model_gini.fit(x_train, y_train)
```

```python
# Prediction and Computing accuracy
pred = model_gini.predict(x_test)
np.mean(pred==y_test)
```

### Decision Tree Regression example

```python
# Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
```

```python
array = iris.values
X = array[:, 0:3]
y = array[:, 3]
```

```python
y
```
